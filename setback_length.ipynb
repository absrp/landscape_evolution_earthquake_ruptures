```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
setback_length = np.array([30,
30,
30,
30,
30,
25,
25,
25,
25,
25,
30,
40,
30,
30,
30,
30,
30,
30,
30,
])

setback_length_meters = setback_length * 0.3048
```

```python
plt.figure(dpi=300,figsize=(8,2))
plt.scatter(setback_length_meters, np.zeros_like(setback_length_meters), 
            color='slategray', s=20, label='Single trace')
plt.scatter(setback_length_meters + 20, np.zeros_like(setback_length_meters + 10), 
            color='cornflowerblue', s=20, label='Localized fault zone')
plt.scatter(setback_length_meters + 75, np.zeros_like(setback_length_meters + 75), 
            color='coral', s=20, label='Distributed fault zone')
plt.scatter(setback_length_meters + 250, np.zeros_like(setback_length_meters + 250), 
            color='lightgray', s=20, label='Distributed fractures')
plt.yticks([])  
plt.xlabel('Setback length (meters)')
plt.xscale('log')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.savefig('Figures/setback_length.png', dpi=300, bbox_inches='tight')
```
